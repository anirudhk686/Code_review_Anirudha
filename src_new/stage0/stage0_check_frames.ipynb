```python
import cv2
import numpy as np
import os
import csv
import numpy as np
import pandas as pd
```

```python
label_file =  "../dataset/2/data3D.csv"
```

```python
def csv_reader_jarvis3d(file_name, num_keypoints):
    """
    Return keypoints as numpy array [frames, num_keypoints, 3]
    """
    all_keypoints = []
    with open(file_name) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0
        for row in csv_reader:
            if line_count == 0 or line_count == 1:
                #print(f'{", ".join(row)}')
                line_count += 1
            else:
                keypoints = [float(x) for x in row]
                keypoints = np.asarray(keypoints)
                keypoints = keypoints.reshape([num_keypoints, 4])[:,:4]
                #keypoints = np.delete(keypoints,site_index_to_delete,0)
                all_keypoints.append(keypoints)
                line_count += 1
    all_keypoints = np.asarray(all_keypoints)
    return all_keypoints

keypoints = csv_reader_jarvis3d(label_file, 20)

```

```python
good_frames = []
for i in range(keypoints.shape[0]):
    frame_data = keypoints[i]
    if frame_data[:,3].mean() > 0.7 and frame_data[:,2].mean() < 500:
        good_frames.append(i)
       
```

```python
def split_consecutive_above_threshold(nums, threshold):
    if not nums:
        return []
    
    nums = sorted(nums)  # Sort the array if not already sorted
    result = []
    current_group = [nums[0]]

    for i in range(1, len(nums)):
        if nums[i] == nums[i - 1] + 1:
            # If the current number is consecutive, add it to the current group
            current_group.append(nums[i])
        else:
            # If the group is larger than the threshold, store it
            if len(current_group) > threshold:
                result.append(current_group)
            # Start a new group
            current_group = [nums[i]]
    
    # Don't forget to check and add the last group if it meets the threshold
    if len(current_group) > threshold:
        result.append(current_group)
    
    return result


```

```python
result = split_consecutive_above_threshold(good_frames, 150)
```

```python
result_array = []
i=0
for r in result:
    result_array.append([len(r),r[0], r[-1]])
    i+=1
df = pd.DataFrame(result_array, columns=[ 'length', 'start', 'end'])
df
```
